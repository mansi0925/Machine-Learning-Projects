### Univariate Analysis

```python
data.columns
```

```python
data1 = data[["numcol","yieldpercol","totalprod","totalprod","priceperlb","year"]]
```

```python
data1.head()
```

```python
plt.figure(figsize = (10,10))
plotnumber = 1

for i in data1:
    plt.subplot(3,2,plotnumber)
    sns.histplot(x=data1[i],kde=True)
    plotnumber = plotnumber + 1
plt.tight_layout()
```
