Sometimes we have two or more series and we have to find all those elements that are present in one series but not in other. We can do this easily, using the **Bitwise NOT operator** along with the **pandas.isin() function.**

13. Program to create 2 integer Series and find all those elements of series 1 that are not present in series 2

```python
# Creating 2 pandas Series
ps1 = pd.Series([2, 4, 8, 20, 10, 47, 99])
ps2 = pd.Series([1, 3, 6, 4, 10, 99, 50])
  
print("Series1:")
print(ps1)
print("\nSeries2:")
print(ps2)
  
# Using Bitwise NOT operator along
# with pandas.isin()
print("\nItems of ps1 not present in ps2:")
res = ps1[~ps1.isin(ps2)]
print(res)
```

39. Write a Pandas program to get the day of month, day of year, week number and day of week from a given series of date strings. 

```python
dateSeries = pd.Series(['09 Feb 2020', '12/01/2020', '21-03-2020', '30 Apr 2020 11:40'])
print(dateSeries)

#parse from dateutil.pareser will help to parse the different formats of date
from dateutil.parser import parse

#map function will return a map object (an iterator) of the results after applying the given function to each item
dateSeries = dateSeries.map(lambda x: parse(x))
print(dateSeries)

print("Day of month", dateSeries.dt.day.tolist())
print("Day of year", dateSeries.dt.dayofyear.tolist())
print("Week number", dateSeries.dt.weekofyear.tolist())
print("Day of week", dateSeries.dt.weekday.tolist())
```

join() - It is a method to combine two differently indexed dataframes into a single dataframe. This is quite similar to the “merge” operation, except the joining operation will be on the “index” instead of  the “columns”.

```python
df1 = pd.DataFrame({"Int_Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2001, 2002,2003,2004])
 
df2 = pd.DataFrame({"Low_Tier_HPI":[50,45,67,34],"Unemployment":[1,3,5,6]}, index=[2001, 2003,2004,2004])
 
print("1st Dataframe: \n ", df1)
print("\n")
print("2nd Dataframe: \n ", df2)
print("\n")
joined= df1.join(df2)
print("The joined dataframe: \n", joined)
```
