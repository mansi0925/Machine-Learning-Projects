# NumPy Operations

## Arithmetic

You can perform array operations with array arithmetic, or scalar with array arithmetic. Let's see some examples:

```python
import numpy as np
arr = np.arange(0,20)
arr
```

```python
arr + arr
```

```python
arr * arr
```

```python
arr - arr
```

```python
# Warning on division by zero, but not an error!
# Just replaced with nan-->not a number
arr/arr
```

```python
# Also warning, but not an error instead infinity(Reciprocal)
1/arr
```

```python
arr
```

```python
arr**3 ## power of operator
```

## Universal Array Functions

Numpy comes with many [universal array functions](http://docs.scipy.org/doc/numpy/reference/ufuncs.html), which are essentially just mathematical operations you can use to perform the operation across the array. Let's show some common ones:

```python
#Taking Square Roots
np.sqrt(arr)
```

```python
#Calcualting exponential (e^)
np.exp(arr)
```

```python
np.max(arr) #same as arr.max()
```

```python
np.sin(arr)
```

```python
np.log(arr)
```

```python
## conversion from degree to radians
d=np.arange(0,120,30).reshape(2,2)
d
```

```python
rad=d*np.pi/180
```

```python
rad
```

```python
np.Sec(rad)
```
