# Python Operators

## Arithmetic Operators

```python
a=5
b=2
x='Data'
y ='Science'

```

```python
#  Addition
c = a + b
print('Addition  of  {}  and  {}  will  give  :-  {}\n'.format(a,b,c))
```

```python
#Concatenate string using plus operator
z = x+y
print(z)

```

```python
# Subtraction
c = a - b
print('Subtracting  {}  from  {}  will  give  :-  {}\n'.format(b,a,c))

```

```python
# Multiplication
c = a * b
print('Multiplying  {}  and  {}  will  give  :-  {}\n'.format(a,b,c))

```

```python
# Division
c = a / b
print('Dividing  {}  by  {}  will  give  :-  {}\n'.format(a,b,c))

```

```python
# Modulo of both number
c = a % b
print('Modulo  of  {}  ,  {}  will  give  :-  {}\n'.format(a,b,c))

```

```python
# Power
c = a ** b
print('{}  raised  to  the  power  {}  will  give  :-  {}\n'.format(a,b,c))

```

```python
# Division(floor)
c = a // b
print('Floor  division  of  {}  by  {}  will  give  :-  {}\n'.format(a,b,c))
```

## Logical Operators

```python
x = True
y = False

print('Logical  AND  operation  :-  ',x  and  y)  #  True  if  both  values  are  true
print('Logical  OR  operation  :-  ',x  or  y)  #  True  if  either  of  the  values  is  true
print('NOT  operation  :-  ',not  y  )  #  True  if  operand  is  false
```

# If  conditional statement

* A Python if statement evaluates whether a condition is equal to true or false.
* The statement will execute a block of code if a specified condition is equal to true. Otherwise, the block of code within the if statement is not executed

```python
x=2
if x==2: # If condition is true
    print('The if condition is true')
```

```python
x=5
if x==2: # If condition is false
    print('The if condition is true')
```

```python
x=5
if x==2: # If condition is false
    print('The if condition is true')
else:
    print('Else part is executed')
```

```python
x=-23
if(x>0):
    print('i am in if block')
else:
    print('i am not')
```

```python
if(5<4):
    print('5 is big')
    print('i am in if block')



print('i am out of if block')
```

```python
x = 5

if x>0:
    print("X is positive")
    print("i am in IF block")

print("i am  out of IF block")
```

```python
# if ... else

x = -5

if x>0:
    print("X is positive")
    print("i am in IF block")
else:
    print("X is Negative")
    print('i am in else block')

print("i am  out of IF block")
```

```python
x=5
if(x>10):
    print('i am in if block')
elif(x<6):   # elif will help in checking more than 1 if condition
    print('i am in elif block')
else:
    print('i am in else block')
```

```python
# if... elif ... else
x =int(input('enter the number'))

if x>0:
    print("X is positive")
elif x<0:
    print("X is Negative")
    print('i am in elif block')
else:
    print("X is Zero")
    print('i am in else block')

print('outside of the loop')
```

```python
## Nesting
x=2
if(x>=3):
    print('i am in 1st if')
    if(x>2):
        print('i am in 2nd if')
        if(x==4):
            print('i am in 3rd if ')
```

```python
i=2
if(i>0):
    print('first loop')
    if(i<=5):
        print('secound loop')
        print('number is in primary range')
        if(i>5):
            print('mid range')
        else:
            print('end of the loop')
else:
    print('number is negative')

print('out of the nested loop')
```

```python
i = 10
if (i == 10):

    #  First if statement
    if (i < 15):
        print("i is smaller than 15")

    # Nested - if statement
    # Will only be executed if statement above
    # it is true
    if (i < 12):
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
```

# Dictionary

Dictionary is a mutable data type in Python.
A python dictionary is a collection of key and value pairs separated by a colon (:) & enclosed in curly braces {}.
Keys must be unique in a dictionary, duplicate values are allowed
