## Hypothesis Testing

```python
from scipy import stats
```

```python
# Experiment :- what is the average age of male marrying in INDIA

# 30 is the average marriage age of male in India?

# hypothesised_mean = 30

# H0 = Marriage age of male in INDIA is 30
# Ha = Marriage age of male in INDIA is not 30


data = [32,28,29,29,31,32,33,32,35,37]

stats.ttest_1samp(data,30)
```

```python
## Rule of thumb

# if p value is less than 5%, we reject the null hypothesis
# if p value is greater than 5%, we fail to reject null hypothesis

## A p value is value telling how much your null hypothesis is correct

# p value = 0.07 = 7% > 5% , we failed to reject null hypothesis, we accept the fact marriage age of male in india is 30
```

```python
## Marriage age of male in India is similar to marriage age of male in US?
## Two sample independent test

mar_age_ind = [32,28,25,29,31,32,33,32,35,32,37]
mar_age_us  = [34,36,39,40,37,35,38,42,38,28,29]
```

```python
# H0 = marriage age of male in india is similar to marriage age of male in USA
# Ha = marriage age of male in india is not similar to marriage age of male in USA

stats.ttest_ind(mar_age_ind, mar_age_us)
```

```python
stats.ttest_ind(mar_age_ind, mar_age_ind)
```

```python
# P value is 0.011 == 1.1% < 5% -- we reject the null hypothesis, we accept alternative hypothesis
# As the statistics value is in negative, it average marriage age of male in INDIA is less than average marriage age of male
# in US
```

```python
## Company redbull claims that it gives you energy
# Two sample dependent 

before_rb = [45,53,66,42,57]
after_rb  = [35,42,35,46,32]
```

```python
# H0 = redbull has the same effect before and after the drinks
# Ha = redbull has not the same Effect before and after the drinks

stats.ttest_rel(before_rb,after_rb)
```

```python
## pvalue is 0.07 == 7% > 5% == we fail to reject the null hypothesis, we accept the fact that redbull donot gives us energy
```

```python
# ANOVA
## Analysis of variance

lemon= [36,12,46,10,15,20]
coffee = [33,10,38,12,13,19]
rb  = [31,9,39,13,10,15]
```

```python
# H0 : All the drinks are having same effect
# Ha : All the drinks are not having the same effect

stats.f_oneway(lemon,coffee,rb)
```

```python
## p value is 0.885 = 88.5% > 5%, we fail to reject the null hypothesis, we have accept the fact all the drinks have the same
## effect over the human body
```
