```python
data.isnull().sum()
```

```python
# check distribution of albumin and globulin ratio
sns.histplot(data=data,x="Albumin_and_Globulin_Ratio",kde=True)
#since data is left skewed , we handle missing values with median
```

```python
 # Check median for albumin and globulin ration column
data['Albumin_and_Globulin_Ratio'].median()
```

```python
# Replace the nan values in albumin and globulin Ratio column with median value
data.loc[data['Albumin_and_Globulin_Ratio'].isnull()==True,'Albumin_and_Globulin_Ratio']=0.93
```

```python
plt.figure(figsize=(10,15), facecolor='white')
plotnumber = 1

for column in data1:
    if plotnumber<=9 :
        ax = plt.subplot(3,3,plotnumber)
        sns.boxplot(x=data1[column])
        plt.xlabel(column,fontsize=10)
        plt.ylabel('count',fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# How to find outliers in total bilirubin
Q1_tb=data['Total_Bilirubin'].quantile(0.25)
print("lower quartile",Q1_tb)
Q3_tb=data['Total_Bilirubin'].quantile(0.75)
print('upper quartile',Q3_tb)
```

```python
# Find IQR
IQR_tb=Q3_tb-Q1_tb
IQR_tb
```

```python
# Find Lower limit & upper limit
lower_limit_TB=Q1_tb-1.5*IQR_tb
print('lower limit is',lower_limit_TB)
upper_limit_TB=Q3_tb+1.5*IQR_tb
print('upper limit is',upper_limit_TB)
```

```python
#Find values above upper_limit
data.loc[data['Total_Bilirubin']>upper_limit_TB]
```

```python
# find percentage of outliers
len(data.loc[data['Total_Bilirubin']>upper_limit_TB])/len(data)*100
```

```python
# find percentage of outliers
len(data.loc[data['Total_Bilirubin']<lower_limit_TB])/len(data)*100
```
