## Bar Plot

Bar plot represents the category of data with rectangular bars with length and heights propotional to values that represent

```python
data = {'C':20, 'Java':30, 'C++':15, 'Python':45}
```

```python
courses = list(data.keys())
courses
```

```python
values = list(data.values())
values
```

```python
courses = ['C', 'Java', 'C++', 'Python']
values = [20, 30, 15, 45]
```

```python
plt.bar(courses, values)
```

```python
plt.figure(figsize = (5,5))

plt.bar(courses, values, color = 'green', width = 0.5)
plt.xlabel('Courses Offered')
plt.ylabel('Number of students enrolled')
plt.title('Students enrolled in different cources')
plt.show()
```

```python
plt.figure(figsize = (5,5))

plt.bar(courses, values, color = ['green', 'red','blue', 'yellow'], width = 0.5)
plt.xlabel('Courses Offered')
plt.ylabel('Number of students enrolled')
plt.title('Students enrolled in different cources')
plt.show()
```

```python
plt.figure(figsize = (5,5))

plt.bar(courses, values, color = ['green', 'red','blue', 'yellow'], edgecolor= 'black', width = 0.5)
plt.xlabel('Courses Offered')
plt.ylabel('Number of students enrolled')
plt.title('Students enrolled in different cources')
plt.show()
```

## Hist Plot

```python
Bloodpressure = [110,111,113,115,112,115,118,119,120,123,125,128,126,132,100]
```

```python
plt.hist(Bloodpressure, bins = [103,106,109,112,115,118,121,124,127,130,133,136])
plt.xlabel('Bloodpressure')
plt.ylabel('Frequency')
plt.title('Bloodpressure vs Frequency')
plt.show()
```

## Pie Plot

```python
Fruits = ["Mango", "Banana", "Apple", "Kiwi"]
Numbers = [10, 20, 12, 7]
```

```python
plt.pie(Numbers, labels = Fruits)
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct = '%.1f')
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct = '%.1f', explode = [0, 0.1, 0, 0.2])
plt.show()
```

```python
plt.pie(Numbers, labels = Fruits, autopct = '%.1f', explode = [0, 0.1, 0, 0.2], colors = ['Green', 'Yellow', 'Red', 'Blue'])
plt.show()
```
