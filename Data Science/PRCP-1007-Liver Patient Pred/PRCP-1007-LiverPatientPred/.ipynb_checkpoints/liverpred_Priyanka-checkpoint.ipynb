## DOMAIN ANALYSIS


*   Age: indicates person age upto 90
*   Sex:column contain 441 male patient records and 142 female patient records.
*   Total Bilirubin- It is a combination of direct and indirect bilirubin.
*   Direct Bilirubin-
*   Alkaline Phosphotase:Alkaline phosphatase (ALP) is an enzyme that's found throughout your body. ALP blood tests measure the level of ALP in your blood that comes from your liver and bones, and it's one of the tests included in a comprehensive metabolic panel.
*   Alanine Aminotransferase:It is an enzyme found mostly in the liver. An ALT test measures the amount of ALT in the blood. When liver cells are damaged, they release ALT into the bloodstream. High levels of ALT in your blood may be a sign of a liver injury or disease.
*   Aspartate Aminotransferase:The aspartate aminotransferase (AST) test is a blood test that checks for liver damage.
*   Total Protein:The normal range is 6.0 to 8.3 grams per deciliter (g/dL) or 60 to 83 g/L.
*   Albumin:An albumin blood test checks your liver and kidney function. Albumin is protein in your blood plasma. Low albumin levels might be the result of kidney disease, liver disease, inflammation or infections. High albumin levels are usually the result of dehydration or severe diarrhea.
*   Albumin and Globulin Ratio:An A/G ratio test measures albumin and globulin, the two main proteins in your blood. This simple blood test is used to monitor your nutritional status, immune function, and overall health. High or low A/G ratios are particularly associated with kidney and liver disease.
*   Target: it denotes 1 : patient with  liver disease and 2: patient with no liver disease disease

## Dropping feature which are more correlated to target.

1. Direct Bilirubin: we drop these feature because total bilirubin contain direct and indirect bilirubin,so we drop this.

2. Asparate Aminotransferase: it is blood test to find liver damage, in alamine aminoterferase also done blood test to find liver damage, so we drop one.

3. Albumin: it is protein test. In albumin and globulin ratio we find ratio albumin and globulin, so we drop albumin feature.

```python
data.drop(['direct_bilirubin', 'aspartate_aminotransferase', 'albumin'], axis="columns", inplace=True)
```

```python
## Split data into x and y
x = data.drop('target', axis=1)
y = data.target
```

```python
## training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# training accuracy
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,classification_report,f1_score
```

```python
precision_score(y_test,y_pred)## out of predicted result, actual true positive result
```

```python
recall_score(y_test,y_pred)## out of predicted result , true positive rate
```

```python
accuracy_score(y_test,y_pred)## accuracy (comparsion of predicted and test values)
```

```python
confusion_matrix(y_test,y_pred) ## to see true positive,false negative,false positive and true negative
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    "criterion":("gini", "entropy"),
    "splitter":("best", "random"),
    "max_depth":(list(range(1, 20))),
    "min_samples_split":[2, 3, 4],
    "min_samples_leaf":list(range(1, 20)),
}
```

```python
tree_clf = DecisionTreeClassifier(random_state=3)
tree_cv = GridSearchCV(tree_clf, params, scoring="f1", n_jobs=-1, verbose=1, cv=3)
tree_cv.fit(x_train,y_train)
best_params = tree_cv.best_params_
print(f"Best paramters: {best_params})")
```

```python
tree_cv.best_score_
```

```python
model=DecisionTreeClassifier(criterion='gini',max_depth=1,min_samples_leaf= 1,min_samples_split=2,splitter='best')
model.fit(x_train,y_train)#training the model
y_predhp=model.predict(x_test)#prediction
```

```python
f1_score(y_test,y_predhp)
```

```python
accuracy_score(y_test,y_predhp)
```

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
```

```python
#dictionary for hyperparameters
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_clf1 = RandomForestClassifier(random_state=10)#model

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='f1',param_distributions=random_grid, n_iter=50, cv=3,
                               verbose=2, random_state=10, n_jobs=-1)
rf_cv.fit(x_train, y_train)
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")
```

```python
rf_clf2 = RandomForestClassifier(n_estimators= 2000, min_samples_split=2, min_samples_leaf= 2, max_features='auto', max_depth=100, bootstrap=True)#passing best parameter to randomforest
rf_clf2.fit(x_train,y_train)
y_predRFHP=rf_clf2.predict(x_test)
```

```python
f1_score(y_test,y_predRFHP)
```
